```python
# Import SparkSession to start Spark
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("BigDataAnalysis").getOrCreate()

print("Spark Started Successfully")
```

```python
# Import pandas to download dataset from URL
import pandas as pd

# Dataset URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

# Load dataset into pandas DataFrame
pdf = pd.read_csv(url)

# Save dataset locally as CSV
pdf.to_csv("tips.csv", index=False)
```

```python
# Load CSV file into PySpark DataFrame
df = spark.read.csv("tips.csv", header=True, inferSchema=True)

# Display first few rows of dataset
df.show()
```

```python
# Print schema (structure of dataset)
df.printSchema()
```

```python
# Count total number of records (rows)
df.count()
```

```python
from pyspark.sql.functions import avg

# Calculate average total bill
df.select(avg("total_bill")).show()
```

```python
# Group data by gender and calculate average total bill
df.groupBy("sex").avg("total_bill").show()
```

```python
# Group data by day and calculate average total bill
df.groupBy("day").avg("total_bill").show()
```

```python
# Group data by day and calculate average tip
df.groupBy("day").avg("tip").show()
```
