```python
# Import required libraries
from textblob import TextBlob   # For sentiment analysis
import pandas as pd             # For handling data
```

```python
# Create a sample dataset of reviews (text data)
data = {
    "review": [
        "The food was amazing",
        "Service was very bad",
        "I love this place",
        "Not good experience",
        "Average food",
        "Excellent service and tasty food"
    ]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

```python
# Define a function to analyze sentiment of each review
def get_sentiment(text):
    # Create TextBlob object
    analysis = TextBlob(text)

    # Check polarity value
    # Polarity > 0 → Positive
    # Polarity < 0 → Negative
    # Polarity = 0 → Neutral

    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Apply sentiment function to each review
df["Sentiment"] = df["review"].apply(get_sentiment)

# Display updated dataset with sentiments
df
```

```python
# Count number of Positive, Negative and Neutral reviews
sentiment_count = df["Sentiment"].value_counts()

# Display result
sentiment_count
```

```python
# (Optional but good for visualization)

import matplotlib.pyplot as plt

# Plot bar chart of sentiment counts
sentiment_count.plot(kind='bar')

# Add title and labels
plt.title("Sentiment Analysis Result")
plt.xlabel("Sentiment")
plt.ylabel("Count")

# Show plot
plt.show()
```
